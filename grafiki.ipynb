Цель: визуализировать данные из датасета в виде графиков. А именно: построить с помощью библиотек matplotlib и pandas линейные графики (минимум три штуки), столбчатые диаграммы (минимум три штуки), круговые диаграммы (минимум две штуки), диаграммы рассеяния (минимум две штуки), гистограммы (минимум три штуки). Датасет из прошлой практической был заменён, так как там почти не было числовых данных (заменён с вашим разрешением).


```python
import pandas as pn
import numpy as np
import matplotlib.pyplot as plt

steam = pn.read_csv('game_data_all.csv')
steam
```

```python
plt.figure(figsize=(25,5))
plt.plot(steam['game'].head(10),steam['peak_players'].head(10),label = 'Пиковое число игроков',color = 'yellow', marker = 'o')
plt.title('Статистика пикового значения игроков')
plt.xlabel('Название игры')
plt.ylabel('Число пикового значения игроков')
plt.grid()
plt.legend()
plt.show()
```

Линейный график статистики пикового значения игроков. Было взято 10 игр из датасета. Можно сделать вывод, что из всех 10 игр наибольшее количество игроков было в Resident Evil 4.
График строился с помощью библиотеки matplotlib.pyplot. steam['game'].head(10) и steam['peak_players'].head(10) берёт первые 10 значений из начала датасета. Label задает название
величины для легенды. Color это цвет графика, marker это точки на графике. plt.xlabel и plt.ylabel это названия осей x и y. plt.legend показывает легенду графика. plt.show показывает сам график. plt.title показывает название графика. plt.figure используется для создания фигуры, plt.grid используется для показа сетки.

```python
plt.figure(figsize=(25,5))
plt.plot(steam['game'].head(10),steam['total_reviews'].head(10),label = 'Все отзывы',color = 'red', marker = 'o')
plt.plot(steam['game'].head(10),steam['positive_reviews'].head(10),label = 'Положительные отзывы',color = 'blue', marker = 'o')
plt.title('Статистика всех отзывов и положительных отзывов')
plt.xlabel('Название игры')
plt.ylabel('Отзывы')
plt.grid()
plt.legend()
plt.show()
```

Линейный график статистики всех отзывов и позитивных отзывов. Было взято также 10 игр из начала датасета и соотношения всех отзывов к игре и позитивных отзывов к игре. Можно сделать вывод, что количество положительных отзывов почти совпадает с количеством всех отзывов, из чего делаем вывод, что негативных отзывов к этим играм мало. Было построено два графика с помощью plt.plot и данных внутри, color это цвет графика, marker это точки на графике, plt.grid используется для показа сетки.

```python
plt.figure(figsize=(25,5))
plt.plot(steam['game'].head(10),steam['total_reviews'].head(10),label = 'Все отзывы',color = 'red', marker = 'o')
plt.plot(steam['game'].head(10),steam['negative_reviews'].head(10),label = 'Негативные отзывы',color = 'blue', marker = 'o')
plt.title('Статистика всех отзывов и негативных отзывов')
plt.xlabel('Название игры')
plt.ylabel('Отзывы')
plt.grid()
plt.legend()
plt.show()
```

Линейный график статистики всех отзывов и негативных отзывов. Было взято также 10 игр из начала датасета и соотношения всех отзывов к игре и негативных отзывов к игре. Можно сделать вывод, что количество негативных отзывов по сравнению со всеми отзывами очень мало, следовательно позитивных отзывов больше. Было построено два графика с помощью plt.plot и данных внутри, color это цвет графика, marker это точки на графике, plt.grid используется для показа сетки.

```python
plt.figure(figsize=(27,5))
plt.bar(steam['game'].head(10),steam['all_time_peak'].head(10),label = 'Самое пиковое значение играющих за всё время',color = 'orange')
plt.xlabel('Название игры')
plt.ylabel('Пиковое значение играющих за всё время')
plt.title('Статистика пикового значения играющих за всё время')
plt.legend()
plt.show()
```

Столбчатая диаграмма самого пикового значения играющих за всё время. Было взято также 10 игр из начала датасета и значение всего пикового количества играющих за всё время. Можно сделать вывод, что наиболее часто запускаемая игра была Resident Evil 4. Столбчатая диаграмма была построена с помощью plt.bar. Метод создания похож на создание линейного графика, но только без маркера.

```python
plt.figure(figsize=(35,5))
plt.bar(steam['game'].head(12),steam['rating'].head(12),label = 'Рейтинг',color = 'orange')
plt.xlabel('Название игры')
plt.ylabel('Значение рейтинга')
plt.title('Статистика рейтинга игр')
plt.legend()
plt.show()
```

Столбчатая диаграмма статистики рейтинга. Было взято 12 игр и рейтинг. Можно сделать вывод, что у этих игр достаточно большой рейтинг и они почти равны. Столбчатая диаграмма была построена с помощью plt.bar. Метод создания похож на создание линейного графика, но только без маркера.

```python
plt.figure(figsize=(35,5))
plt.bar(steam['game'].head(12),steam['total_reviews'].head(12),label = 'Все отзывы',color = 'orange')
plt.bar(steam['game'].head(12),steam['positive_reviews'].head(12),label = 'Положительные отзывы',color = 'blue')
plt.xlabel('Название игры')
plt.ylabel('Отзывы')
plt.title('Соотношение всех отзывов и положительных отзывов')
plt.legend()
plt.show()
```

Столбчатая диаграмма соотношения положительных отзывов и всех отзывов. В выводе можно сказать, что количество положительных отзывов почти совпадает с количеством всех отзывов.

```python
genres_count = steam['primary_genre'].value_counts().head(10)
plt.figure(figsize=(15, 8))
plt.pie(genres_count, labels=genres_count.index, autopct='%1.1f%%')
plt.title('Топ 10 игровых жанров в стиме')
plt.show()
```

Круговая диаграмма показывает распределение топ-10 жанров игр в стиме. Каждый сектор представляет собой определенный жанр, а процент внутри сектора указывает на долю этого жанра от общего числа игр. Можно сделать вывод, что наиболее популярным жанром является Indie, так как у него самый большой процент. Строится с помощью plt.pie.

```python
top_developers = steam['developer'].value_counts().head(10)
plt.figure(figsize=(10, 5))
plt.pie(top_developers, labels=top_developers.index, autopct='%1.1f%%')
plt.title('Топ разработчиков в стиме')
plt.show()
```

Круговая диаграмма показывает соотношение топовых разработчиков на платформе Steam. Каждый сектор представляет собой одного из топовых разработчиков, а процент внутри сектора указывает на их долю от общего числа игр. Можем сделать вывод, что наиболее популярным разработчиком является Choice of Games, так как у него самый большой процент. Строится с помощью plt.pie.

```python
plt.figure(figsize=(5, 3))
plt.scatter(steam['positive_reviews'], steam['negative_reviews'], alpha=0.5)
plt.title('Диаграмма рассеивания: Положительные и Отрицательные отзывы')
plt.xlabel('Положительные отзывы')
plt.ylabel('Отрицательные отзывы')
plt.show()
```

Диаграмма рассеивания положительных и отрицательных отзывов. Можно сделать вывод, что отзывы независимы между собой. Диграмма была построена с помощью plt.scatter.

```python
plt.figure(figsize=(5, 4))
plt.scatter(steam['peak_players'], steam['positive_reviews'], alpha=0.5)
plt.title('Диаграмма рассеивания между пиковым количеством игроков и положительными отзывами')
plt.xlabel('Пиковое количество игроков')
plt.ylabel('Количество положительных отзывов')
plt.show()
```

Диаграмма рассеивания положительных и отрицательных отзывов. Можно сделать вывод, что явной зависимости между пиковым количеством игроков и положительными отзывами нет. Диграмма была построена с помощью plt.scatter.


```python
genres_counts = steam['primary_genre'].value_counts()
plt.figure(figsize=(6, 3))
plt.hist(genres_counts, bins=10, color='blue', edgecolor='black', label = 'Распределение жанров')
plt.title('Гистограмма частоты жанров')
plt.xlabel('Жанр')
plt.ylabel('Частота')
plt.legend()
plt.show()
```

Гистограмма была построена с помощью plt.hist из 10 ячеек. Можно сделать вывод, что один из жанров очень преобладает, другие же менеет преобладающие.

```python
plt.figure(figsize=(10, 6))
plt.hist(steam['rating'], bins=20, color='blue', edgecolor='black',label = 'Распределение рейтинга')
plt.title('Гистограмма рейтинга')
plt.xlabel('Рейтинг')
plt.ylabel('Частота')
plt.legend()
plt.show()
```

Гистограмма была построена с помощью plt.hist, было взято 20 ячеек. Можно сделать вывод, что самый часто встречающийся рейтинг находится между 60 и 80, то есть около 70.

```python
plt.figure(figsize=(10, 6))
plt.hist(steam['review_percentage'], bins=20, color='blue', edgecolor='black', label = 'Процент отзывов')
plt.title('Гистограмма соотношения отзывов')
plt.xlabel('Процент отзывов')
plt.ylabel('Частота')
plt.legend()
plt.show()
```

Гистограмма была построена с помощью plt.hist, было взято 20 ячеек. Можно сделать вывод, что самый часто встречающийся процент положительных отзывов находится между 80 и 100, то есть около 90 и он встречается чаще всего.
